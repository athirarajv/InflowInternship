## 1. Show the number of 'confirmed', 'deaths' and 'recovered' cases in each region(use groupby function).

```python
a = df.groupby(['Country/Region', 'Confirmed', 'Deaths', 'Recovered / 100 Cases'])
```

```python
a.first()
```

## 2. Remove all the records where 'Confirmed' cases are less than ten(< 10) [use ~ ].

```python
b = df.loc[~df['Confirmed'] < 10]
```

```python
b
```

## 3. In which region were maximum number of confirmed cases recorded? 

```python
c = df[df.Confirmed == df.Confirmed.max()]
```

```python
c
```

```python
c[['Country/Region', 'Confirmed']]
```

## 4. In which region were mimimum number of deaths recorded?

```python
d = df[df.Deaths == df.Deaths.min()]
```

```python
d
```

```python
d[['Country/Region', 'Deaths']]
```
